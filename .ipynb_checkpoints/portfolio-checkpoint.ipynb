```python
prices_df = yf.download(tickers, 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
prices_df.head()
```

```python
close_df = prices_df['Adj Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
close_df = close_df.reindex(all_weekdays)
close_df = close_df.fillna(method='ffill')
```

```python
d05 = close_df.loc[:, 'D05.SI']

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_d05 = d05.rolling(window=20).mean()
medium_rolling_d05 = d05.rolling(window=50).mean()
long_rolling_d05 = d05.rolling(window=100).mean()

# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(d05.index, d05, label='D05.SI')
ax.plot(short_rolling_d05.index, short_rolling_d05, label='20 days rolling')
ax.plot(medium_rolling_d05.index, medium_rolling_d05, label='50 days rolling')
ax.plot(long_rolling_d05.index, long_rolling_d05, label='100 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
plt.show()
```

```python
class TrackPortfolio:
    
    def __init__(self, ticker, indicators, start_date, end_date):
        self._ticker = ticker
        self._indicators = indicators
        self._start_date = start_date
        self._end_date = end_date

    def dl_data(self):
        df = yf.download(self._ticker, 
                              start=self._start_date, 
                              end=self._end_date, 
                              progress=False)
        df = ta.add_all_ta_features(
            df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
        all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
        df = df.reindex(all_weekdays)
        self._df = df.fillna(method='ffill')
        
    def plot_chart(self):
        fig, ax = plt.subplots(figsize=(16,9))
        for i in self._indicators:
            j = self._df.loc[:,i]
            ax.plot(j.index, j, label= i)
        ax.set_xlabel('Date')
        ax.set_ylabel(self._ticker)
        ax.legend()
        plt.show()
```
